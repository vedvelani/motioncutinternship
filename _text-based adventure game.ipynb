```python
import time

def introduction():
    print("Welcome to the Text Adventure Game!")
    time.sleep(1)
    print("You find yourself in a dark, mysterious cave.")
    time.sleep(1)
    print("Your goal is to find the treasure and make it out alive.")

def choose_path():
    print("\nChoose your path:")
    time.sleep(1)
    print("1. Go left")
    time.sleep(1)
    print("2. Go right")
    time.sleep(1)

    choice = input("Enter your choice (1 or 2): ")

    if choice == '1':
        return 'left'
    elif choice == '2':
        return 'right'
    else:
        print("Invalid choice. Try again.")
        return choose_path()

def encounter_monster():
    print("\nOh no! You've encountered a fierce monster!")
    time.sleep(1)
    print("What will you do?")
    time.sleep(1)
    print("1. Fight the monster")
    time.sleep(1)
    print("2. Run away")

    choice = input("Enter your choice (1 or 2): ")

    if choice == '1':
        print("You bravely fight the monster, but it's too strong.")
        time.sleep(1)
        print("Game over. Better luck next time!")
        return 'game_over'
    elif choice == '2':
        print("You wisely choose to run away.")
        time.sleep(1)
        return 'left'
    else:
        print("Invalid choice. Try again.")
        return encounter_monster()

def find_treasure():
    print("\nCongratulations! You've found the hidden treasure.")
    time.sleep(1)
    print("You make your way out of the cave and emerge victorious!")

def play_game():
    introduction()
    current_location = choose_path()

    while True:
        if current_location == 'left':
            current_location = encounter_monster()
            if current_location == 'game_over':
                break
        elif current_location == 'right':
            find_treasure()
            break

if __name__ == "__main__":
    play_game()
```
