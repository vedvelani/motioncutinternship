```python
class Task:
    def __init__(self, description, due_date, priority):
        self.description = description
        self.due_date = due_date
        self.priority = priority
        self.completed = False

class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, description, due_date, priority):
        task = Task(description, due_date, priority)
        self.tasks.append(task)

    def display_tasks(self):
        if not self.tasks:
            print("No tasks in the to-do list.")
        else:
            print("To-Do List:")
            for i, task in enumerate(self.tasks, start=1):
                print(f"{i}. Description: {task.description}")
                print(f"   Due Date: {task.due_date}")
                print(f"   Priority: {task.priority}")
                print(f"   Completed: {'Yes' if task.completed else 'No'}")
                print()

    def complete_task(self, task_index):
        if 0 < task_index <= len(self.tasks):
            self.tasks[task_index - 1].completed = True
            print("Task marked as completed.")
        else:
            print("Invalid task number. Please try again.")

    def update_task(self, task_index, description, due_date, priority):
        if 0 < task_index <= len(self.tasks):
            task = self.tasks[task_index - 1]
            task.description = description
            task.due_date = due_date
            task.priority = priority
            print("Task updated.")
        else:
            print("Invalid task number. Please try again.")

    def remove_task(self, task_index):
        if 0 < task_index <= len(self.tasks):
            removed_task = self.tasks.pop(task_index - 1)
            print(f"Task '{removed_task.description}' removed.")
        else:
            print("Invalid task number. Please try again.")

def main():
    todo_list = ToDoList()

    while True:
        print("\nTo-Do List Menu:")
        print("1. Add Task")
        print("2. Display Tasks")
        print("3. Mark Task as Completed")
        print("4. Update Task")
        print("5. Remove Task")
        print("6. Quit")
        choice = input("Enter your choice: ")

        if choice == "1":
            description = input("Enter task description: ")
            due_date = input("Enter due date (optional): ")
            priority = input("Enter priority (Low/Medium/High): ")
            todo_list.add_task(description, due_date, priority)
            print("Task added to the to-do list.")
        elif choice == "2":
            todo_list.display_tasks()
        elif choice == "3":
            task_index = int(input("Enter the task number to mark as completed: "))
            todo_list.complete_task(task_index)
        elif choice == "4":
            task_index = int(input("Enter the task number to update: "))
            description = input("Enter new task description: ")
            due_date = input("Enter new due date (optional): ")
            priority = input("Enter new priority (Low/Medium/High): ")
            todo_list.update_task(task_index, description, due_date, priority)
        elif choice == "5":
            task_index = int(input("Enter the task number to remove: "))
            todo_list.remove_task(task_index)
        elif choice == "6":
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
